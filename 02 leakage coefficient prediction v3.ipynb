**SOTTOINSIEME DATASET DI INTERESSE (PEZZO BUONO, VELOCITA_1=140rpm, VELOCITA_A_REGIME=2300rpm, TEMPERATURA=40°C)**

```python
df = df.query("n_esito == 100 and velicita_1 == 140 and velocita_a_regime == 2300" + 
              "and Temperatura >= 39 and Temperatura <= 43" +
              "and media_portata_velocita_a_regime > 0 and media_pressione_velocita_a_regime > 0" +
              "and media_portata_velocita_1 > 0 and media_pressione_velocita_1 > 0")
df.shape
```

**RICERCA E GESTIONE MISSING VALUES**

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=' ')
df = pd.DataFrame(imp.fit_transform(df), columns=df.columns)
sum(df.isnull().sum())
```

**RIMOZIONE OUTLIERS UNIVARIATI**

```python
from scipy.stats import zscore
for var in interested_vars:
    if (var != "n_esito" and df[var].std() != 0):
        df['zscore'] = pd.Series(np.abs(zscore(df[var])))
        idx = list(df.query("zscore > 3.5").index)
        print(var, len(idx))
        df.drop(idx, inplace=True)
        df.drop("zscore", 1, inplace=True)
```

```python
for col in interested_vars:
    df[col] = df[col].astype(float)
df.shape
```

**CALCOLO DEL LEAKAGE COEFFICIENT IPOTIZZANDO UN'EFFICIENZA NOTA A REGIME (non conosco la portata teorica)**

```python
df['eff_regime'] = 0.95
```

alfa_140 = alfa_regime <br>
1: (portata_teorica_140 - portata_140) / pressione_140 = (portata_teorica_regime - portata_regime) / pressione_regime <br> <br>
eff_regime = portata_regime / portata_teorica_regime --> portata_teorica_regime = portata_regime / eff_regime <br>
eff_140 = portata_140 / portata_teorica_140          --> portata_teorica_140 = portata_140 / eff_140 <br>
k = eff_140 / eff_2300 (spiegazione https://www.machinerylubrication.com/Read/28430/hydraulic-pump-motors-maintenance) <br> <br>
sostituendo in 1 e sviluppando: <br>
k = A / (1 + eff_regime * (A - 1)), dove A = (portata_140 * pressione_regime) / (pressione_140 * portata_regime) <br>
eff_140 = k * eff_regime <br>
portata_teorica_140 = portata_teorica / eff_140 <br>
per ogni riga, si ottiene efficienza e portata teorica a 140 rpm (conoscendo efficienza a regime fissa) <br>

A questo punto sappiamo tutto per calcolare alfa_140 che coincide con alfa_regime (dipende da caratteristiche interne) <br>
Ciò che cambia da 140 rpm a 2300 rpm è l'efficienza volumetrica non il coefficiente di perdita <br>
alfa_140 = (portata_teorica_140 - portata_140) / pressione_140 <br>
alfa_regime = (portata_teorica_regime - portata_regime) / pressione_regime <br>

```python
A = (df.media_portata_velocita_1 * df.media_pressione_velocita_a_regime) / (df.media_pressione_velocita_1 * df.media_portata_velocita_a_regime)
df['eff_1'] = (df.eff_regime * A) / (1 + df.eff_regime * (A - 1)) 
df.eff_1.describe()
```

```python
df['portata_teorica_velocita_1'] = df.media_portata_velocita_1 / df.eff_1
df.portata_teorica_velocita_1.describe()
```

```python
df['portata_teorica_velocita_a_regime'] = df.media_portata_velocita_a_regime / df.eff_regime
df.portata_teorica_velocita_a_regime.describe()
```

```python
alfa_1 = (df.portata_teorica_velocita_1 - df.media_portata_velocita_1) / df.media_pressione_velocita_1
alfa_1.describe()
```

```python
alfa_regime = (df.portata_teorica_velocita_a_regime - df.media_portata_velocita_a_regime) / df.media_pressione_velocita_a_regime
alfa_regime.describe()
```

**A QUESTO PUNTO SAPPIAMO ALFA E POSSIAMO SVILUPPARE ALGORITMI DI REGRESSIONE O CLUSTERING PER LA PREDIZIONE**

```python
df['alfa'] = alfa_1
df.to_csv(os.path.join(path, "final_data.csv"), index_label=False)
```
