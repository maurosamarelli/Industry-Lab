```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=' ')
df = pd.DataFrame(imp.fit_transform(df), columns=df.columns)
sum(df.isnull().sum())
```

**RIMOZIONE OUTLIERS UNIVARIATI**

```python
from scipy.stats import zscore
for var in interested_vars:
    if (var != "n_esito" and df[var].std() != 0):
        df['zscore'] = pd.Series(np.abs(zscore(df[var])))
        idx = list(df.query("zscore > 3.5").index)
        print(var, len(idx))
        df.drop(idx, inplace=True)
        df.drop("zscore", 1, inplace=True)
```

```python
for col in interested_vars:
    df[col] = df[col].astype(float)
```

**QUESTO E' UN ESEMPIO PER UN METODO ALTERNATIVO MA NON SO SE E' CORRETTO** <br><br>
La portata teorica **Qt** si ottiene moltiplicando lo spostamento volumetrico **Vd** per la velocitÃ  **N** --> **Qt = Vd * N** <br>
Indichiamo **Qa**:portata ottenuta, **p**:pressione, **alfa**:leakage coefficient <br><br>
Dai risultati sperimentali sappiamo che: **Qa = Qt - alfa * p**, con velocitÃ =140rpm e temperatura=40Â°C <br>
Supponendo (o verificando?) anche che: **Qa = Qt - alfa * p**, con velocita=2300rpm e temperatura=40Â°C <br>
e che il leakage coefficient dipenda solo dalla struttura interna quindi non cambia a diverse velocitÃ : <br><br>
E' possibile impostare il seguente sistema di due equazioni in due variabili: <br>
1: **Qa@140rpm  = Qt@140rpm  - alfa * p@140rpm** <br>
2: **Qa@2300rpm = Qt@2300rpm - alfa * p@2300rpm** <br><br>
1: **Qa@140rpm  = Vd * 140rpm  - alfa * p@140rpm** <br>
2: **Qa@2300rpm = Vd * 2300rpm - alfa * p@2300rpm** <br><br>
Le due variabili incognite sono: **Vd** e **alfa** (leakage coefficient) <br><br>
Risolvendo il sistema si ha: <br>
**alfa = (Qa@2300rpm - k * Qa@140rpm) / (k * p@140rpm - p@2300rpm)**, dove k = 2300/140 = 16.43

```python
k = 2300 / 140
df['alfa_'] = (df.media_portata_velocita_a_regime - k * df.media_portata_velocita_1) / (k * df.media_pressione_velocita_1 - df.media_pressione_velocita_a_regime)
df.alfa_.describe()
```
