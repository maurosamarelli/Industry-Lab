**SOTTOINSIEME DATASET DI INTERESSE (PEZZO BUONO, VELOCITA_1=140rpm, VELOCITA_A_REGIME=2300rpm) - CALCOLO ALPHA**

```python
filters = "n_esito == 100 and velicita_1 == 140 and velocita_a_regime == 2300 and media_portata_velocita_a_regime > 0 and media_pressione_velocita_a_regime > 0 and media_portata_velocita_1 > 0 and media_pressione_velocita_1 > 0"
df = df.query(filters)
df['portata_teorica_velocita_1'] = 108.36
df['alpha'] = (df.portata_teorica_velocita_1 - df.media_portata_velocita_1) / df.media_pressione_velocita_1
interested_vars = ['Programma', 'media_coppia_finale', 'media_pressione_velocita_1', 'media_portata_velocita_1', 'portata_teorica_velocita_1', 'alpha', 'media_pressione_velocita_a_regime', 'media_portata_velocita_a_regime', 'Temperatura']
df = df[interested_vars]
df.shape
```

**RICERCA MISSING VALUES**

```python
for var in interested_vars:
    print(var, df[var].isnull().sum())
```

**ANALISI DIFFERENZIATA PER PROGRAMMA (DIVERSI MODELLI DI GP5)**

```python
groups = df.Programma.value_counts()
groups
```

```python
main_programs = list(groups.where(lambda x : x > 1000).dropna().index)
gp5 = {}
for p in main_programs:
    gp5[p] = df.query("Programma == '" + p + "'")
    gp5[p].drop("Programma", 1, inplace=True)
    print(p, gp5[p].shape)
```

**RIMOZIONE OUTLIERS UNIVARIATI**

```python
from scipy.stats import zscore
for p in main_programs:
    print(p)
    print("Before", gp5[p].shape[0])
    for column in gp5[p].columns:
        gp5[p][column] = gp5[p][column].astype(float)
        z = np.abs(zscore(gp5[p][column]))
        gp5[p] = gp5[p][z <= 3.5]
    print("After", gp5[p].shape[0], "\n")
```

```python
num_vars = ['media_pressione_velocita_a_regime', 'media_portata_velocita_a_regime', 'media_pressione_velocita_1', 'media_portata_velocita_1', 'alpha', 'Temperatura']
sns.set(color_codes=True)
fig, axes = plt.subplots(figsize=(20, 10), nrows=int(len(num_vars)/2+0.5), ncols=2)
for i, column in enumerate(num_vars):
    sns.distplot(gp5[main_programs[0]][column], ax=axes[i//2, i % 2])
```

**BOX PLOT DISTRIBUZIONE ALPHA PER OGNI GRUPPO**

```python
data = []
for p in main_programs:
    data.append(gp5[p].alpha)
fig, ax = plt.subplots(figsize=(20,8))
ax.boxplot(data, meanline=True, showmeans=True, meanprops = dict(linestyle='--', linewidth=3))
ax.set_title("Alpha")
plt.xticks(list(1 + np.arange(7)), main_programs)
plt.show()
```

**CORRELAZIONE TRA LE GRANDEZZE PER OGNI GRUPPO**

```python
program = main_programs[0] #you can change
print(program)
round(gp5[program][num_vars].corr(), 3)
```
